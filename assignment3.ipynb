#### Chart - 15 - Pair Plot

````python
# Chart - 15 visualization code

# Selecting important numerical columns
selected_cols = ['iyear', 'nkill', 'nwound']

# Pair Plot
plt.figure(figsize=(10,6))
sns.pairplot(df[selected_cols], diag_kind='kde')

plt.suptitle("Pair Plot of Year, Kills, and Wounded", y=1.02)
plt.show()
````

##### 1. Why did you pick the specific chart?

A pair plot is used to visualize pairwise relationships between multiple numerical variables simultaneously. It provides both scatter plots for relationships and distribution plots for individual variables, making it highly effective for multivariate analysis.
In this case, variables like year (iyear), number of kills (nkill), and number of wounded (nwound) are selected to understand their interactions and patterns.

##### 2. What is/are the insight(s) found from the chart?

There is a strong positive relationship between number of kills and number of wounded, indicating that as the severity of an attack increases, both fatalities and injuries rise proportionally.
The distribution plots show that both nkill and nwound are highly right-skewed, meaning most attacks result in low casualties, but a few incidents cause extremely high damage.
The relationship between year and casualties is not strictly linear, but certain clusters suggest periods of intensified terrorist activity.
Presence of outliers indicates rare but extremely impactful attacks, which significantly influence overall trends.
